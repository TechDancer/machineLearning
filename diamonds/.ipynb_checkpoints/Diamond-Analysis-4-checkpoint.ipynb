## Normalize/Scale the other input columns

```python
from sklearn import preprocessing
                                            
caratDF = diamonds['carat'].to_frame()
print (caratDF.tail())
print (caratDF.shape)

scaler = preprocessing.StandardScaler().fit(caratDF)
caratDF = pd.DataFrame(scaler.transform(caratDF))
print (caratDF.tail())
print ('\nscaled/normalized version has some problem, why is 15684 negative ???')
diamonds.rename(columns={'Unnamed: 0':'caratScaled'}, inplace=True)
diamonds['caratScaled'] = caratDF 

diamonds.loc[(diamonds['carat'] > 2)].sample(n=20)
               
```

```python
# cut out other columns
diamonds = diamonds.drop(columns=['caratScaled','cut','color','clarity','depth','table','x','y','z'])
print(diamonds.shape)
diamonds.head()
```
